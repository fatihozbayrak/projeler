## Kütüphaneleri import edin

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import matplotlib as mpl# gorsellestirmeleri yapmak icin kullaniyoruz

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 

import warnings;
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m=np.arange(0,11.)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize =(6.3,5))
plt.plot(m,E,color="b" ,linewidth=6)
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title(label="E=mc^2" );
```

```python
plt.figure(figsize =(8,5))
plt.plot(m,E,color="orange" ,linewidth=3.5)
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title(label="E=mc^2" )
plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
E = np.log(m)*c**2
E

plt.figure(figsize =(6,5))
plt.plot(m,E,color="g" ,linewidth=3.5)
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.yscale('log')
plt.title(label="E=mc^2" )
plt.grid(which="minor")
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize =(8,7))
plt.plot(labels,july16_2007,color="royalblue",linewidth=2, label="july16_2007")# x ve y degiskenlerini belirttigimiz degiskenlerle beraber ciz diyoruz
plt.plot(labels,july16_2020,color="darkorange",linewidth=2, label="july16_2020")
plt.legend() 
plt.show()
```

```python
plt.figure(figsize =(8,5))
plt.plot(labels,july16_2007,color="cornflowerblue",linewidth=1.5
         , label="july16_2007")# x ve y degiskenlerini belirttigimiz degiskenlerle beraber ciz diyoruz
plt.plot(labels,july16_2020,color="darkorange",linewidth=2.5, label="july16_2020")
plt.legend(bbox_to_anchor=(1.25, 0.55, 0., 0.)) 
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig,ax =plt.subplots(2,1, figsize=(10,7)) 

ax[0].plot(labels,july16_2007,color="royalblue",linewidth=4)
ax[0].set_title("july16_2007" )

ax[1].plot(labels,july16_2020,color="royalblue",linewidth=4)
ax[1].set_title("july16_2020" )

plt.tight_layout();
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python

fig,ax = plt.subplots(figsize=(12,7))
ax.plot(labels, july16_2007,linewidth=3, color="blue")
ax.set_ylabel("july16_2007", color="blue", fontsize=14)
ax2=ax.twinx()
ax2.plot(july16_2020,linewidth=3,color="red")
ax2.set_ylabel("july16_2020",color="blue",fontsize=14)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
ax.spines['right'].set_color('red')
ax.spines['left'].set_color('blue')
ax.spines['right'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)

ax.plot(labels, july16_2007,linewidth=3, color="blue")

ax.set_ylabel("july16_2007",color="blue", fontsize=14)
ax2=ax.twinx()
ax2.plot(july16_2020,linewidth=3,color="red")
ax2.set_ylabel("july16_2020",color="blue",fontsize=14);
```
